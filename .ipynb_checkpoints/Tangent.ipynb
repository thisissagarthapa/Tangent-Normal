Tangent and normal  Dervative questions

1.Finds the points on the curve where the tangent are parallel to the x-axis  and equation  of curve is 4y=x^4-8x^2.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 400)
y = (x**4 - 8*x**2) / 4

x_points = np.array([0, 2, -2])
y_points = (x_points**4 - 8*x_points**2) / 4

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', label='Curve: 4y = x^4 - 8x^2')
plt.plot(x_points, y_points, 'ro', label='Points where tangent is parallel to x-axis')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the Curve 4y = x^4 - 8x^2')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

2.At what angle does the curve y(1+x)=x cut the x-axis?

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)
y = x / (1 + x)

slope = 1
intercept = 0  # Tangent line passes through (0,0)

x_tangent = np.linspace(-2, 2, 400)
y_tangent = slope * x_tangent + intercept

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', label='Curve: $y(1 + x) = x$')
plt.plot(x_tangent, y_tangent, 'r--', label='Tangent at $(0, 0)$')

plt.plot(0, 0, 'go', label='Intersection Point $(0, 0)$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve $y(1 + x) = x$ and Tangent Line at $(0, 0)$')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.show()
```
